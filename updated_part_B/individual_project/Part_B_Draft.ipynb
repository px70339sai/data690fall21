```python
# Considering data from the year 2000
poverty_headcount_2000 = poverty_headcount.query("Year == 2000")

poverty_headcount_2000 .sample(5)
```

```python
poverty_headcount_2000 = poverty_headcount_2000.query("indicator in ['1.0.HCount.2.5usd']")

poverty_headcount_2000
```

```python
df_pivot = poverty_headcount_2000.pivot_table(index=['Country Code','Country Name','Region','Income Group','Lending Type'], columns= 'indicator').reset_index()
df_pivot.head()
```

# Poverty Headcount (2.50$ Per Day)

```python
fig = px.pie(
    df_pivot,
    values='value', 
    names='Country Name',
    title='Poverty Headcount',
    hover_name='Country Code',
)
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```
